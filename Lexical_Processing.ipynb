### Introduction

This section covers **word frequency analysis** as a fundamental step in text analytics. Analyzing word frequencies helps in understanding the structure, patterns, and significance of words in a large text corpus.

Key concepts include:

- **Zipfâ€™s Law**: The frequency of a word is inversely proportional to its rank in a frequency table. A few words occur very frequently while most words occur rarely.
- **Word Categories**:
  - **Stop Words**: Highly frequent, low-information words (e.g., "is", "the") often removed to reduce noise and computational cost.
  - **Significant Words**: Important words that carry meaningful information about the text.
  - **Rare Words**: Infrequent words that may still provide useful information in tasks like spam detection.

**Applications**:
- Preprocessing for machine learning tasks such as spam detection, sentiment analysis, and text classification.
- Stop word removal improves efficiency by reducing dataset size and number of features.
- Word frequency distributions provide insights for exploratory data analysis in NLP.

### Question 2: Remove Stop Words and Find Most Frequent Word

**Description:**  
In this exercise, you will remove **stop words** from the text of a book and then find the frequency of the most frequent remaining word.

**Steps:**  
1. Load the book from the URL.  
2. Split the text into words.  
3. Remove stop words using NLTK's stopword list.  
4. Build a frequency distribution of the remaining words.  
5. Print the frequency of the most frequent word.


```python
# Colab-ready code for Exercises: Word Frequency and Stopwords

# Install nltk if not already installed (uncomment if needed)
# !pip install nltk

import requests
from nltk import FreqDist
from nltk.corpus import stopwords
import nltk

# Download NLTK stopwords
nltk.download('stopwords')

# -----------------------------
# Exercise 1: Third Most Frequent Word
# -----------------------------
# Load the ebook
url = "https://www.gutenberg.org/files/16/16-0.txt"
peter_pan = requests.get(url, verify=False).text

# Split the book into words
peter_pan_words = peter_pan.split()

# Build frequency distribution
word_frequency = FreqDist(peter_pan_words)

# Frequency of third most frequent word
third_freq = word_frequency.most_common(3)[2][1]

print("Third most frequent word frequency:", third_freq)

# -----------------------------
# Exercise 2: Remove Stopwords and Find Most Frequent Word
# -----------------------------
# Get NLTK stopwords list
stop_words = stopwords.words('english')

# Remove stopwords
no_stops = [word for word in peter_pan_words if word.lower() not in stop_words]

# Build frequency distribution for remaining words
word_frequency_no_stops = FreqDist(no_stops)

# Most frequent word's frequency
most_freq_no_stop = word_frequency_no_stops.most_common(1)[0][1]

print("Most frequent word frequency after removing stopwords:", most_freq_no_stop)
```

# ðŸ“Œ Tokenisation
Tokenisation is a fundamental **text preprocessing** technique used in **Natural Language Processing (NLP)**.  
It involves breaking raw text into smaller units called **tokens**, which can be:

- Words
- Characters
- Sentences
- Subwords or symbols

In a **spam detection system**, tokenisation is crucial because:
- Machine Learning models **cannot work directly with text**
- Text must be converted into **numeric features**
- Each token (word) can act as a **feature** for classification

## ðŸš¨ Challenges in Text Data
Real-world text data contains significant **noise**, such as:
- Inconsistent letter casing (UPPER/lower)
- Punctuation marks
- Spelling errors
- Contractions (can't, won't, hasn't)

Simple methods like Pythonâ€™s `split()` fail to handle these complexities properly.

## ðŸ§  Why NLTK Tokenisers?
NLTK tokenisers are more intelligent than basic string splitting because they:
- Correctly handle **contractions**
- Preserve meaningful words like *â€œoâ€™clockâ€*
- Support different types of text (formal, social media, etc.)

## ðŸ” Types of Tokenisation
| Tokeniser Type | Description |
|----------------|------------|
| Word Tokeniser | Splits text into individual words |
| Sentence Tokeniser | Splits text into sentences |
| Character Tokeniser | Splits text into characters |
| Tweet Tokeniser | Handles emojis, hashtags, mentions |
| Regex Tokeniser | Custom tokenisation using regex |

Tokenisation is the **first step** before applying techniques like:
- Stopword removal
- Stemming / Lemmatization
- Vectorization (Bag of Words, TF-IDF)

### Description
Write a piece of code that breaks a given sentence into words and stores them in a list.  
Remove the **stop words** from the list and then print the length of the remaining words.  
Use the **NLTK word tokeniser**.

**Sample Input:**  
â€œEducation is the most powerful weapon that you can use to change the worldâ€

**Expected Output:**  
6


```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import ast, sys

sentence = sys.stdin.read()

# change sentence to lowercase
sentence = sentence.lower()

# tokenise sentence into words
words = word_tokenize(sentence)

# extract nltk stop word list
stopwords = set(stopwords.words('english'))

# remove stop words
no_stops = [word for word in words if word.isalpha() and word not in stopwords]

# print length - don't change the following piece of code
print(len(no_stops))
```

### Description
Use **NLTKâ€™s regex tokeniser** to extract all the mentions from a given tweet.  
A mention starts with `@` followed by alphabets, numbers, or underscores.  
Print the total number of mentions.

**Sample Tweet:**  
So excited to be a part of machine learning and artificial intelligence program made by @upgrad and @iiitb

**Expected Output:**  
2


```python
from nltk.tokenize import regexp_tokenize
import ast, sys

text = sys.stdin.read()

# change text to lowercase
text = text.lower()

# pattern to extract mentions
pattern = r'@\w+'

# extract mentions using regex tokeniser
mentions = regexp_tokenize(text, pattern)

# print length - don't change the following piece of code
print(len(mentions))
```
